```python
num = int(input("Enter a number "))
if num%2 == 0:
    print(num," even number")
else:print(num," odd number")
```

```python
# check sign

def check_sign(num):
    if num > 0:
        print("positive")
        return 1
    elif num < 0:
        print("negative")
        return -1
    else:
        print("zero")
        return 0
    

print(check_sign(num))

```

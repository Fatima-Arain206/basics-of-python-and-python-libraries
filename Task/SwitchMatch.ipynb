```python
n = int(input("Enter number 1 to 7 ")) 
match n:
        case 1:
                print("Monday..")
    
        case 2:
                print("tuesday")   
        case 3:
                print("Wednesday") 
        case 4:
                print("thursday") 
        case 5:
                print(" friday")
        case 6:
                print(" Saturday")   
        case  7: print("sunday")  
        case _:# default
                print(" enter valid number")
                        
                                           

```

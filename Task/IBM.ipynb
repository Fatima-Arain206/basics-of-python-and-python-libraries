```python
weight = float(input(" enter weight in kgs"))
height = float(input("Enter height in meters "))
BMI = weight / height* height
if BMI < 18.5:
    print("Under weight ")
elif BMI >= 18.5 and BMI <=24.5:
    print("normal")    
elif BMI >= 24.5 and BMI <=29.9:
    print("over weight") 
    
elif BMI >=30:
    print("obese")
```

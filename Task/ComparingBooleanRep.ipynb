```python
n = int(input("Enter a number"))
equal = True
bit = int(0)
while(n > 0):
 last = int(n%2)
 bit = bit*10+last
 n=  n//2

print(bit)
```

```python
ans = str(bit)
for i in range(len(ans)-1):
    if ans[i] != ans[i+1]:
        equal = False
        break
if equal:
    print(f" equal {bit}")
else:
    print(f"not equal {bit}")
```

```python
ch = input("enter a char ")
if  len(ch) > 1:
    print("Enter only one character ")
    ch = input()

if ch.isdigit():
    print("your input is a number") 
elif ch.isupper():    
    print(" char is in upper case")
elif ch.islower():
    print("it is lower case")
else:    
    print("char is special char ")    
```

```python
ch = input(" enter a single char")
if len(ch) >1:
    
    ch=input(" single char !!!!!!!!>>>>>>>")
is_digit= False
is_char = False
alphabet ="abcdefghijklmnopqrstuvwxyz"
for char in alphabet:
    if ch.lower() == char or ch.upper() == char:
       # print(" char ..... ")
        is_char = True
        break
digit ="0123456789"
for d in digit:
    if ch == d:
     is_digit = True


if is_digit:
    print(" digit ")
elif is_char:
    print("char")    
else:
    print("special char ",ch)    

```

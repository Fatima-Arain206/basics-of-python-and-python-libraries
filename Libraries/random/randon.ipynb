```python
def score_counter(score_list):
    negative_scores = 0
    neutral_scores = 0
    positive_scores = 0

    for score in score_list:
        if score >= 9:
            positive_scores += 1
        elif score >= 6:
            neutral_scores += 1
        else:
            negative_scores += 1

    print('Negative:', negative_scores)
    print('Neutral:', neutral_scores)
    print('Positive:', positive_scores)
```

Agar aap random.seed(100) likhein gi, to wo save point badal jaye ga, aur ab koi naya pattern hamesha fix ho jaye ga (for example: 2, 8, 3, ...).

##range

range(1, 11): It is like a compressed (.zip) file. It does not create all numbers instantly; it only holds the formula to save memory.

list(): It acts as an extractor. It opens that zip file and converts the hidden numbers into an actual visible list like [1, 2, 3, ..., 10].

range(1, 4) directly in a variable: If you don't use list(), the file remains zipped. That's why Python just prints range(1, 4) instead of showing the actual numbers 1, 2, 3.

random.seed(99): It does NOT set the range from 1 to 99. It is just a "Save Point" or a "Channel Number" to keep the random results exactly the same every time you run the code.

my_var = range(1, 5): Printing this will NOT show [1, 2, 3, 4]. It will only print range(1, 5) because the data is still compressed (zipped). To see the actual numbers, you must extract it using list(range(1, 5)).

random.choice() vs random.choices(): random.choice() picks only one item, while random.choices() (with an 's') can pick multiple items at once and returns them as a new list.

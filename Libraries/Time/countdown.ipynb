```python
import time
set_time=int(input("enter time in seconds"))

for x in range(set_time):
    time.sleep(1)
    print(x)
print("times up")
```

```python
# down
def countdown_timer(set_time):
    for y in range(set_time,0,-1):
        seconds = y % 60
        minutes = int( y / 60) % 60
        hours = int(y / 3600)
        print(f"{hours:02}:{minutes :02}:{seconds:02}")
        time.sleep(1)

    print("time is over!")
countdown_timer(3605)
```

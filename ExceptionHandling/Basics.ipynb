```python
def divide_number(a,b):
    try:
        result=a/b
        print(f"{a} / {b} ={round(result)}")
    except ZeroDivisionError:
        print(f"{b} is undefined divisor")
    except TypeError:
        print(f" there is a datatype  mistake")
    finally:
        print(f"how  was that ?? cool na ")
divide_number('8',0)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 5x5 pixels ki image, har pixel mein 3 rang (RGB) hain
# 255 = Full brightness, 0 = No color

pixel_array = np.array([
    [[255, 0, 0],   [0, 255, 0],   [0, 0, 255],   [255, 255, 0], [255, 0, 255]],  # Row 1
    [[0, 255, 255], [0, 0, 0],     [255, 255, 255],[0, 0, 0],     [0, 255, 255]],  # Row 2
    [[255, 0, 255], [255, 255, 255],[255, 0, 0],   [255, 255, 255],[255, 0, 255]],  # Row 3
    [[0, 255, 255], [0, 0, 0],     [255, 255, 255],[0, 0, 0],     [0, 255, 255]],  # Row 4
    [[255, 0, 255], [255, 255, 0], [0, 0, 255],   [0, 255, 0],   [255, 0, 0]]   # Row 5
])

# Is array ko screen par image ki tarah show karne ke liye:
plt.imshow(pixel_array)
plt.axis("off") # Grid lines chupane ke liye
plt.show()
```

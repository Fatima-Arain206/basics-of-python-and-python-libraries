```python
''' add all even numbers from a list to b'''

def filter_all_even(list):
    try:
        even=[]
        for x in list:
            if x % 2==0:
                even.append(x)
         
    except TypeError:
        print("enter a list ")
    except ValueError:
        print (" enter  only number list ")
    except IndexError:
        print(" enter a valid list")
    finally:
        print("list is filtered ...")
    return even

filter_all_even([24,5,6,8,9,5,3,2,3,4,5])
        
```

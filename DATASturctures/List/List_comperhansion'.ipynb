```python
# right rotation
def right_rotation(a:list,r):
    '''this method will right-rotate n times of given  if list = [1,2,3] and r=1 output= [3,1,2]'''
    try:
        if len(a)==r:
            return a
        # last=a.pop(-1)
        for  n in range(r):
            last=a.pop(-1)
            i=-len(a)-1
            while i > 0:
                a[i]= a[i+1]
                i-=1
            a.insert(0,last)
    except Exception as e:
        print({e})
    else:
        return a


right_rotation(['a','b','c','d'],2)
```

```python
# make a new list of common elements
def common(a:list,b:list):
    try:

        com=[]
        for x in a:
            if x in b:
                com.append(x)
    except TypeError:
        print("enter a list ")
    else:
        return com
a=   [1,4,590,6,42,44]
b= [155.3,67,43,90,1,44]
com= [x for x in a if x in b] # list comp

# common([1,4,590,6,42,44],[155.3,67,43,90,1,44])
print(com)
```

```python
# palindrome
def check_palindrome(a):
    try:
        i=0
        j=len(a)-1
        while i < j:
            if a[i] != a[j]:
                return False,a
            i+=1
            j-=1

    except Exception as e:
        print({e})
    else:
        return True
check_palindrome('natan')
```

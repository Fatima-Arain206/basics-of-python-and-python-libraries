```python
def reverse_list(list):
    i=0
    j=len(list)-1
    while i < j:
        temp= list[i]
        list[i]= list[j]
        list[j] = temp
        i+=1
        j-=1
    return list
reverse_list([1,2,3,4,5,6])

```

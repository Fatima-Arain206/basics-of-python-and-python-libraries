```python
# remove duplicate values
def unique_values(a:list):
    try:
        unique_list=list()
        for x in a:
            if x in unique_list:
                continue
            else:
                unique_list.append(x)
    except TypeError:
        print(" enter a valid list")
    except ValueError:
        print({ValueError} ,"  enter good values")
    except Exception as an:
        print({an} ,'error')
    else:
        return unique_list

unique_values([2,3,5,3,2,1,9,7,5,4,3,67,87,6,'f,','f',
                                             ''])
```

```python
#   convert nested list into flat
nested=[[1,2,3,6],[3,4,5,7]]
flat=[]
for x in range(len(nested)):
    for y in range(len(nested[0])):
        flat.append(nested[x][y])
flat


flat
```

```python
# find second largest
def second_largest(a):
    try:
        max_val=float('-inf')
        sec_max=float('-inf')
        for x in a:
            if x > max_val:
                sec_max=max_val
                max_val=x

            if x < max_val and x > sec_max:
                sec_max=max_val
    except Exception as e:
        print({e})
    else:
        return sec_max

second_largest(flat)
```

```python
from json import tool

# shopping cart program
foods=[]
prices = []
total = 0

while True:
    food = input(" enter a food to buy  or (q to quit)")
    if food.lower() == 'q':
        break
    else:
        price = float(input(f" enter the price of {food }:PKR"))
        foods.append(food)
        prices.append(price)

print(f'----- YOUR CART -----')
for fd in range(len(foods)):



    print(" {foods[fd]:>12}, { prices[fd]   }  ")

    total=total+fd
print()

print(f"Total {total}:PKR")

```

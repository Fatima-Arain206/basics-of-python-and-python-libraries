```python
# party lsit
def list_of_guest(name_list,is_coming):
    try:
        i=0
        if len(name_list)!= len(is_coming):
            return -1
        while i < len(name_list):
            if is_coming[i]==False:
                name_list.remove(name_list[i])
            i+=1
        print(name_list)
    except ValueError:
        print("mistmatch")

name_list=['ali',"Fatima","iqra","soohn"]
is_coming=[False,True,True,False]
print(list_of_guest(name_list,is_coming))
```

A tuple is an immutable sequence that can contain elements of any data type.
Tuples are instantiated, or expressed, with parenthesis ​or the tuple() function.
## The position of the element is fixed in tuples, ​so you can't add new elements in the middle, ​and you can't change any of the elements. ​If we try to change the last name, ​which lives in index number 2, from Hopper to Copper, ​the code will throw an error.


```python
# converting a list into tuple
a=[3,4,6,8,0,4,32.1,1,3,]
print(type(a))
a=tuple(a)# tuplse function
print(type(a))
# voice vrs
a=list(a)
print(type(a))
```

```python
def return_tuples(pr):
    '''when a function is returning more then one value it is a tuple'''
    doller=int(pr//1)
    cent=round(pr%1*100)
    return doller,cent

c=return_tuples(9075)
c
print(type(c))
```

Iteration 1 (Pehla Chakkar): Loop team ki list mein se pehla packet ("Ali", 21, "Striker") uthaye ga.

Unpacking: Python automatic is packet ko kholega aur iski teeno values ko aapke diye gaye variables mein sequence wise daal dega:

name ban jayega "Ali"

age ban jayegi 21

position ban jayegi "Striker"

```python
def unpacking_team(team):
    re=list()
    for name,age,pos in team:
        re.append('Name: {:<10} age: {:<10} surname: {:<10}'.format(name,age,pos))
    return re


for x in unpacking_team(team):
    print(x)



```

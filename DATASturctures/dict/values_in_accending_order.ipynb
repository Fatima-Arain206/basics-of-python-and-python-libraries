```python
def ascending_order(a):
    try:
        list_key_v= list(a.items())
        print(list_key_v)
        for out in range(len(list_key_v)):
            for j in range(0,len(list_key_v)- out-1):
                if list_key_v[j][1] >= list_key_v[j+1][1]:
                     list_key_v[j], list_key_v[j+1] = list_key_v[j+1],list_key_v[j]


        a={ x[0]:x[1] for x in list_key_v}



    except KeyError:
        print(f" key is invalid")
    except ValueError :
        print(f" values are invalid ")
    else:
        print(f' asscending order')
        return a

a={
    'a':9,
    'b':19,'c':4,'d':5,'e':19,'f':2,'g':8,'h':7,'z':9,
}
ascending_order(a)
```

calculate sum of all integers values in dic

```python
def sum_(di):
    try :
        sum=0
        for x in di:
            if type(di[x])==int:

                sum=sum+di[x]
    except Exception as e:
        print({e})

    else:
        return sum
di={
    'a':22,
    'b':"s",
    'c':23,
    'd':'7894',
    'f':873.73

}

sum_(di)
```

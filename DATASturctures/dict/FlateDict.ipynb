```python
def flate_dict(fl,nested,paresnt_key='',):
    try:

        for x,y in nested.items():
            last_key=(paresnt_key+'_'+x) if paresnt_key else x
            if type(y)==dict:
                flate_dict(fl,y,last_key)
            elif paresnt_key != '':
                fl.update({paresnt_key+'_'+x: y})
            else:
                fl.update({x:y})

    except TypeError:
        print("must enter dict")
    else:
        return fl


nested={
    'a':'hi',
   'aa':{'fatima':'arain','cgpa':3.6},
    'b':{'aqsa':'ali','cgpa':3.3},
   'c':{'nagma':'nizamani','cgpa':3.63},
    'd':{'zees':'raj','cgpa':3.90},
    'f':{'un':{'90':'79','h':'low'}}

}
fl=dict()
print(flate_dict(fl,nested))
```

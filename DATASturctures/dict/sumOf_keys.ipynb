```python
def sumOf_(dic):
    try:
        sum=0
        for x in dic:
            sum=sum+x
    except ValueError:
        print(f"Enter keys in int ")
    except TypeError:
        print(f"Enter keys in int ")
    else:
        return sum
a={11: 1, 2: 2, 23: 1, 3: 2, 0: 'd', 7: 2, 4: 1, 5: 1}

sumOf_(a)
```

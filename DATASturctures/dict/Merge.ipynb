```python
# frequency of char
def fri_st_dic(st):
    try:
        fi={}
        i=0
        visit=[]
        while i < len(st):

            if st[i] in visit:
                i+=1
                continue
            count=1
            j=i+1
            while j < len(st)-1:
                if st[i] == st[j]:

                    count+=1

                j+=1
            fi.update({st[i]:count})
            visit.append(st[i])
            i+=1




    except ValueError:
        print("Enter  string ")
    except TypeError :
        print(TypeError)
    else:
        return fi

fri_st_dic("apple")
```

```python
# same question with diffrent method
def fri_count(st):
    fri={}
    for x in st:
        if x in fri:
            fri[x]+=1
        else:
            fri[x]=1
    return fri
fri_count("fatima")
```

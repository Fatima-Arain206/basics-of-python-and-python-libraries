```python
# swap a key into value and value into keys from a dictionary

def swap_key_value(dic):
    try:
        swap=dict()
        for x in dic:
            key=x
            val=dic[x]
            swap.update({val:key})
        del dic

    except KeyError:
        print(f"Try to inter proper unique keys")
    except ValueError:
        print(f" try to enter unique values \n because this is a swap function ")
    else:
        print('swaped')
        return swap

a={
    1:'fatima',
    2:2,
    6:'surhan',
    4:'kainat'
}
swap_key_value(a)
```

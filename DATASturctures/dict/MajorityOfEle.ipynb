```python
maj=[11,2,23,2,3,6,7,4,3,5,7,6,6,6]
def majority(majority_list):
    try:

        majority_dict= dict()
        for x in majority_list:
            maj=1
            if x in majority_dict:

                majority_dict[x]=majority_dict[x]+1

            else:
                maj=1
                majority_dict[x]=maj
    except ValueError:
        print(f"enter a valid list  ")
    else:return majority_dict

majority(maj)
```

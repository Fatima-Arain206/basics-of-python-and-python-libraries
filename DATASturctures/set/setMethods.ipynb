```python
a={1,3,7,8,0,8,6,54,2,21,85,4}
b={1,90,76,4,53,462,15436,85}
# difference minus a from b
print(a.difference(b))
```

```python
## symmetric difference
''' not common any element in both sets '''
c={1,2,3,4,0,5,6}# common elements will  ignore not count even once
d={0,8,9,7,99,6}
print(c.symmetric_difference(d))
```

```python
# disjoint
'''should not common a single element'''
d={11,8,9,7,99,60}
c={1,2,3,4,0,5,6}
print(c.isdisjoint(d))
```

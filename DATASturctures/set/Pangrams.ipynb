```python
def unique_words(s1,s2):
    try:
        sp1=list(s1.split())
        set1=set(sp1)
        sp2=list(s2.split())
        set2=set(sp2)
        return set1.intersection(set2)
    except Exception as e:
        return e

s1="my name is fatima "
s2='iam fatima '
unique_words(s1,s2)
```

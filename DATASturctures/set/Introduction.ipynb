set is a data structure in Python вЂ‹that contains only unordered, non-interchangeable elements. вЂ‹Sets are instantiated with the set() function вЂ‹or non-empty braces
Each set element is unique and immutable. вЂ‹However, the set itself is mutable. вЂ‹Sets are valuable when storing mixed data вЂ‹in a single row, or a record, in a data table.
вЂ‹They're also frequently used when storing a lot of elements, вЂ‹and you want to be certain that each one вЂ‹is only present once. вЂ‹Because sets are mutable, вЂ‹they cannot be used as keys in a dictionary. вЂ‹There are two ways to create a set. вЂ‹The first way is with the set function. вЂ‹The set function takes an iterable as an argument вЂ‹and returns a new set object. вЂ‹Let's examine the behavior of sets вЂ‹by passing lists, tuples, and strings вЂ‹through the set function. вЂ‹To turn the list вЂ‹containing "foo, bar, baz, and foo" into a set, вЂ‹pass a list through the set function.
вЂ‹And notice, the list loses the second foo. вЂ‹As I've mentioned, each element must be unique in sets. вЂ‹Pass a tuple through the set function вЂ‹using two sets of parentheses: one to tell the computer вЂ‹that the data we are working with is a tuple; вЂ‹the other because the set function вЂ‹only takes a single argument. вЂ‹Again, the same result, вЂ‹only one foo element can be present in the set. вЂ‹Finally, pass a string through the set function. вЂ‹It doesn't return the string, вЂ‹just the singular occurrence of the letters in the string, вЂ‹O and F, in an unordered way. вЂ‹This is because the set function accepts a single argument, вЂ‹and that argument must be iterable.
вЂ‹A string is iterable, so the set function splits it вЂ‹into individual characters and keeps only the unique ones. вЂ‹The second way to instantiate a set is with braces. вЂ‹However, you have to put something inside the braces. вЂ‹Otherwise, the computer will interpret your empty braces вЂ‹as a dictionary. вЂ‹Note here that instantiating a set with braces вЂ‹treats what's inside the braces as literals. вЂ‹So when instantiating a set вЂ‹of only a single string using braces, вЂ‹it returns a set with a single element, вЂ‹and the element is the string itself. вЂ‹Remember, to define an empty set or a new set, вЂ‹it is best to use the set function.
вЂ‹You can only use curly braces when the set is not empty вЂ‹and you are assigning the set to a variable. вЂ‹Also, keep in mind that because the elements вЂ‹inside a set are immutable, вЂ‹a set cannot be indexed or sliced. вЂ‹Now, let's discuss some additional functions вЂ‹you can use on sets. вЂ‹First, intersection finds the elements вЂ‹that two sets have in common. вЂ‹Union finds all the elements from both sets. вЂ‹Difference finds the elements present in one set, вЂ‹but not the other. вЂ‹And symmetric difference finds elements from both sets вЂ‹that are mutually not present in the other.
вЂ‹Python provides built-in methods вЂ‹for performing each of these functions. вЂ‹Let's start with the intersection method, вЂ‹denoted by the ampersand. вЂ‹First, define two sets. вЂ‹Then, apply the intersection function, вЂ‹either by attaching the intersection method to set one, вЂ‹and passing set two to the method's argument, вЂ‹or by using the ampersand operator. вЂ‹Great. Now, let's apply the union function. вЂ‹Use braces to define the two sets, X one and X two.
вЂ‹The goal is to observe where they overlap. вЂ‹Print them using the union method on the X two variable вЂ‹or the union operator symbol. вЂ‹Union is a commutative operation in mathematics, вЂ‹so the overlapping values will be the same вЂ‹no matter what order you put your variables in. вЂ‹The difference operation on sets, however, вЂ‹is not a commutative operation. вЂ‹Just like in math, if you subtract five from seven, вЂ‹you get a different result вЂ‹than if you subtract seven from five. вЂ‹You can use either the difference method вЂ‹or the minus sign as a set operator. вЂ‹Subtracting set two from set one вЂ‹gives us only the elements in set one вЂ‹that are not shared with set two.
вЂ‹But we don't know if set two contains any elements вЂ‹that are not shared with set one. вЂ‹The inverse is also true: subtracting set one from set two вЂ‹gives us only the elements in set two вЂ‹that are not shared with set one. вЂ‹But we don't know if set one contains any elements вЂ‹that are not shared with set two. вЂ‹To get around this and observe the difference вЂ‹between two sets mutually, вЂ‹use the symmetric difference function. вЂ‹As you might have guessed, вЂ‹you can use the symmetric difference method вЂ‹or the symmetric difference operator, expressed by a caret. вЂ‹Symmetric difference outputs all the elements вЂ‹that the two sets do not share. вЂ‹Excellent work with sets in Python!

```python
def group_anagram(list_words):
    try:
        anagram_dict=dict()
        for x in list_words:
            sorted_word=''.join(sorted(x))
            if sorted_word in anagram_dict:
                anagram_dict[sorted_word].append(x)
            else:
                anagram_dict[sorted_word]=[x]
    except IndexError:
        print(IndexError)
    # except Exception as e:
    #     print({e})
    else:
        return anagram_dict

a=['eat','tea','ate','ten','net']
group_anagram(a)
```

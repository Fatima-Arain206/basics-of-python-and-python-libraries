```python
def per(op):
    match op:
        case '+':
            return 1
        case '-':
            return 1
        case '%':
            return 2
        case '/':
            return 2
        case '*':
            return 2
        case '^':
            return 3

print("Enter your equation with spaces between operators and operands")
print("Example: 2 + 3 * 4")
equation = input()

token = equation.split()
fin = int(token[0])

i = 1
storeop = ''
storevar = 0

# FIX: Check if next element exists BEFORE trying to access it
while i < len(token) - 1:  # Stop BEFORE the last element
    op = token[i]
    next_num = int(token[i+1])
    
    if i + 2 < len(token):
        next_op = token[i+2]
        
        if per(op) < per(next_op):
            storeop = op
            storevar = fin
            fin = next_num
            i += 2
            continue
    
    # Do current operation
    if op == '+':
        fin = fin + next_num
    elif op == '-':
        fin = fin - next_num
    elif op == '/':
        fin = fin / next_num
    elif op == '%':
        fin = fin % next_num
    elif op == '*':
        fin = fin * next_num
    else:
        print(f"{op} is not a valid operator")
    
    i += 2

# Apply stored operation
if storeop != '':
    if storeop == '+':
        fin = storevar + fin
    elif storeop == '-':
        fin = storevar - fin
    elif storeop == '/':
        fin = storevar / fin
    elif storeop == '%':
        fin = storevar % fin
    elif storeop == '*':
        fin = storevar * fin
    else:
        print(f"{storeop} is not a valid operator")
if fin  > int(token[0]):
    print(f"{equation} = {fin}")
```

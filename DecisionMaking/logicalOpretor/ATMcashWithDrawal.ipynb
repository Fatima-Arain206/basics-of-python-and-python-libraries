```python
balance = 20000
reqAmount = int(input("enter amount "))
atmhasCash = input("has your atm card cash property (y/n)")

atmhasCash = atmhasCash.upper()
if atmhasCash == "Y":
    atmhasCash = True
else:
    atmhasCash = False
cardBlock = False
if (reqAmount <= balance ) and ( not cardBlock) and atmhasCash:
    print("Enjoy your cash ")
elif cardBlock:
    print(" you can not withdraw your card is blocked ")
elif not atmhasCash :
    print("you can not withdraw your atm has not cash")
```

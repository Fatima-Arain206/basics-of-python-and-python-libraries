```python
temp = int(input("What is temperature"))
powerSavingMode = input("do you want to turn of or on power saving mode (on/off)")
powerSavingMode = powerSavingMode.upper()
if powerSavingMode == 'OFF':
    powerSavingMode= False
else:
    powerSavingMode = True

if  not powerSavingMode  and temp > 25:
    print("beep beep")
    print("AC is turning \n "
          "on close the windows and door")
elif powerSavingMode and temp > 25:
    print(" Turn of power saving mode")
elif temp  < 25 and not powerSavingMode:
    print(" Alert⚠⚠⚠⚠ it is a cool  day ")
    print(" turn on  power saving mode")
```

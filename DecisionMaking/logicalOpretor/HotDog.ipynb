```python
def hotDog(fish,cheese,onion):
    if fish and cheese and onion:
        return (fish+cheese+onion)
    elif fish:
        return int(fish)
    elif cheese:
        return int(cheese)
    elif onion:
        return int(onion)
    else:
        return (fish+cheese+onion)

print(hotDog(False,False,False))
```

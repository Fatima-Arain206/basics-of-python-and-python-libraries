```python
def count_words(sen):

    if len(sen) == 0:
        print("empty")
        return -1
    elif ' ' not in sen:
        print(f" there is not any space in :{sen}")
        return 0
    else:
        count =1
        for i in sen:
            if i == ' ':
                count +=1

        return count

sen = input(" enter anything but take care of spaces")
fi = count_words(sen)
print(f" there are {fi} words in {sen}"  if fi > 0 else "oho" )
```

```python
def repeated_find(st):
    if len(st)==0:
        return -1
    
    for i in range(len(st)):
        unique = True
        for j in range(len(st)):
            if i != j and st[i] == st[j]:
                unique = False
                break
        if unique:
            return st[i]
        
    
print(repeated_find("ffattima"))


```

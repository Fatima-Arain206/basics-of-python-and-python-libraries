```python
def possible_SubStr(st):
    if len(st)>1:

        subList=[]
        i=0
        # for s in st:
        #     subList.append(s)
        while i < len(st):
            j=i
            while j< len(st):
                char = st[i:j+1]

                subList.append(char)
                j=j+1
            i+=1

        return subList
possible_SubStr("fatima")
```

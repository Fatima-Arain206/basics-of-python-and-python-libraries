```python
def exractdigits():
    st = input("Enter a string")
    print("this application will extract all digit ")
    if len(st) ==0:
         print(" empty string  ")

    nst = ""
    for  k in st:
        if k.isdigit():
            continue
        else:
            nst = nst+k

    if st == nst:
        print(f"{st } has not digits")
    else:
        print(f"first string with digits is {st} ")
        print(f"without digit your string is {nst}")

exractdigits()

```

```python
def palindrome(st):

    if len(st)==0:
        print("empty ")
        return -1
    elif len(st)<3:
        print(" this string is to short ")
        return -1
    else:

        ispalindrome = True
        for i in range(len(st)//2):
          j=-1-i
          if st[i] != st[j]:
              ispalindrome = False
              return ispalindrome


        return ispalindrome



palin = input("Enter a string ")
palindrome(palin)
```

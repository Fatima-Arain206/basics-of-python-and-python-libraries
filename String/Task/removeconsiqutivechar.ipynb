```python
def remove_consecutive(st):
    if len (st ) ==0:
        return -1
    reomved =''
    j=0
    while j < len(st)-1 :
        if st[j] != st[j+1]:
            reomved = reomved+st[j]

        j += 1
    reomved +=st[-1]
    return reomved
print(remove_consecutive(""))


```

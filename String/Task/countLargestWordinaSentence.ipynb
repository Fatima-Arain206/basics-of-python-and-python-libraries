```python
def largest_word(sent):
    if len(sent.strip()) == 0:
        return -1
    largest = 0
    word = ''
    count = 0
    current_word = ''
    for i in sent:
        if i == ' ':
            if count > largest:
                largest = count
                word = current_word
            count = 0
            current_word = ''
        else:
            count += 1
            current_word += i
    if count > largest:
        word = current_word
    return word
```

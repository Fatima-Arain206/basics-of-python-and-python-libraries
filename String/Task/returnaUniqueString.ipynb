```python
def uniquest(st):
    if len(st)==0:
        return -1
    visit =[]
    newre =""

    for i in range(len(st)):
        if st[i] in visit:
            # print(i)
            continue
        for j in range(i+1,len(st)):
          if  st[i] ==  st[j]:
              # print(i , j)
              visit.append(st[i])

        newre = newre+ st[i]
    return newre


st = input(" enter a string ")
uni = uniquest(st)
if st =="":
    print(f"empty")
else:
    print(f' your first string was {st }')
    print(f'your unique string is {uni}')
```

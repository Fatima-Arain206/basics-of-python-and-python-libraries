```python
def check_vowels():
 st = input("enter a string  ")
 if len(st) ==0:
     return -1
 st = st.lower()
 i =0
 for j in st:
     if j=='a' or j==  'e' or j=='o'or j=='i'or j=='u':
         i +=1
 print(f"{st} has total {i} vowels")
check_vowels()
```

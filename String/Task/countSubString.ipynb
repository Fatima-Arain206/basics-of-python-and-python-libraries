a main string and sub str is given find that how many times sub string occurs in main

```python
main = ""
sub ='faaa'
def find_noOf_SubStrings(main,sub ):
    if len(main)==0 or len(sub)==0:
        return -1
    count =0
    i=0
    while i<len(main)-len(sub)+1:
        if sub == main[i:i+len(sub)]:
            count+=1
        i+=1
    return count
main = "aaaaa"
sub ='faaa'
find_noOf_SubStrings(main,sub)
```

```python



def check_starting_point():
    main = input(" enter main string....")
    srch = input("enter search string")
    main.lower()
    srch.lower()
    if len(srch) > len(main):
        return -1
    match = True
    for i,j  in enumerate(srch):
        if j!= main[i]:
            match = False
            return match


    return match


print(check_starting_point() )
```

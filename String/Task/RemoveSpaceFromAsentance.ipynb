```python

def deleting_spaces():
    sent = input("write any thing")
    single =""
    if len(sent) ==0:
         print(f"{sent} is empty ")
         return -1
    for i in sent:
        if i == " ":
                continue
        else:
            single = single+i

    if single == sent:
        print(f" there is not space in { sent}")
        return 0
    else:
        print(f"your sentence is:{sent} ")
        print("deleting space xxx")
        print(f"your sentence without space : {single}")
    return single

print(deleting_spaces())
```

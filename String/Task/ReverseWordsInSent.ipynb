```python
def rev_words(sent):
    if len(sent)==0:
        return -1
    word = sent.split()
    i =0
    rev_sent =''

    while i < len(word):
        w = word[i]
        j =0
        while j < len(w):
            rev = w[-1::-1]
            j=j+1
        rev_sent = rev_sent+' '+rev
        i+=1
    return rev_sent






print(rev_words("hello world !"))

```

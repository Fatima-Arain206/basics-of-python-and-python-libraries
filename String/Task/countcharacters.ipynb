in keyword has two roles
in if condition it works as gatekeeper search in entiere list and return true or false it only stop if it finds value in first

```python
m = "fatima"
if 'q' in m:
   print(f"yes{m['q']} ")
else:
    print("no")
```

in loop in keyword work as an iterator

```python
a = input("enter a string ")
a = a.upper()

b = []
for i ,ok in enumerate(a):
    fi =1

    if ok in b:
        continue
    for j in a[i+1:]:
        if j == ok:
            fi +=1

            b.append(j)
    if fi >1:
        print(f"{ok} appears {fi} time in {a}")

```

```python
print(list(enumerate("fatima")))
print(enumerate("fatima"))
```

```python



def precedence(op):
    match op :
        case '+':
            return 1
        case '-':
            return 1
        case '/':
            return 2
        case '//':
            return 2
        case '*':
            return 2
        case '%':
            return 2
        case '^':
            return 3

def string_evalution(eq):


    if not ' ' in eq:
        return "not space "
    token = eq.split()
    first = int(token[0])
    i =0
    store_op =''
    store_val =0
    i=1
    while i < len(token)-1:
        op = token[i]
        next_num=int(token[i+1])

        if i+2 < len(token)-1:
            next_op = token[i+2]
            if precedence(next_op) > precedence     (op):
                store_op = op
                store_val = first
                first= next_num
                i+=2
                continue
        if op=='+':
             first=first+next_num
        elif op == '-':
            first=first-next_num
        elif op == '*':
            first=first*next_num
        elif op == '/':
            first=first/next_num
        elif op == '//':
            first=first//next_num
        else:
            print(f"{op} is not a valid operator")

        i+=2

    if store_op != '':
        if store_val!=0:
            if store_op=='+':
                first=first+store_val
            elif store_op == '-':
                first=first-store_val
            elif store_op == '*':
                first=first*store_val
            elif store_op == '/':
                first=first/store_val
            elif store_op == '//':
                 first=first//store_val
            else:
                print(f"{store_op} is not a valid operator")

    return first
eq = input("equation there should be space after any number and operator")
print(f" {eq} == {string_evalution(eq)}")
```

```python
roman = input("Enter roman number ")
roman = roman.upper()
print(len(roman))
I = 1
V = 5
X = 10
L = 50
C = 100
D = 500
M = 1000
integer = 0
for i in range(len(roman)):
    if roman[i] == 'I':
        value = I
    elif roman[i] == 'V':
        value = V
    elif roman[i] == 'X':
        value = X
    elif roman[i] == 'L':
        value = L
    elif roman[i] == 'C':
        value = C
    elif roman[i] == 'D':
        value = D
    elif roman[i] == 'M':
        value = M
    else:
        print(roman[i], "is not in roman numbers")
        value = 0

    if i < len(roman) -1:
        if roman[i + 1] == 'I':
            next_value = I
        elif roman[i + 1] == 'V':
            next_value = V
        elif roman[i + 1] == 'X':
            next_value = X
        elif roman[i + 1] == 'L':
            next_value = L
        elif roman[i + 1] == 'C':
            next_value = C
        elif roman[i + 1] == 'D':
            next_value = D
        elif roman[i + 1] == 'M':
            next_value = M
        else:
            next_value = 0

        if value < next_value:
            integer -= value
            print(integer)
        else:
            integer += value
    else:
        integer += value

print(f"{roman}=={integer}")


```

Is Code ka Post-Mortem 🔍
Aapke code mein do hissey hain jo and se jure hain: print("Yellow" > "Cyan" and "Brown" > "orange")

Hissa A: "Yellow" > "Cyan"
Python dono ka pehla letter dekhega: 'Y' aur 'C'.

Kya 'Y', ABC mein 'C' ke baad aata hai? Haan!

To ye hissa ho gaya: True.

Hissa B: "Brown" > "orange"
Yahan aik bohot makhsoos coding rule apply hota hai: ASCII Values.

Python ki duniya mein, saare Small Letters (a-z), Capital Letters (A-Z) se bare hote hain.

Iska matlab hai ke chota 'o', bade 'B' se bada mana jayega ('o' > 'B').

Yahan condition check ho rahi hai: Kya bade B wala "Brown", chote o wale "orange" se bara hai? Nahi!

To ye hissa ho gaya: False.

Final Result:
Ab dono hisson ko and ke sath jorein: True and False.
Kyunke and ko dono sides True chahiye hoti hain, is liye final output aayega: False.

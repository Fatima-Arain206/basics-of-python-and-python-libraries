```python
def capitalization(st):
    if len(st) ==0: return -1
    elif len(st)>0:
        cap = ""

        cap = cap+st[0].upper()
        for s in st[1:] :
            char = s

            char = char.lower()
            cap = cap+char

    return  cap

name = " "
name = capitalization(name)
print(name)

```

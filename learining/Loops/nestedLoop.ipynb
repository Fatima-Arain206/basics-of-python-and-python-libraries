```python
for left in range(7):
    for right in range(left,7):
        print(f'[{left}]|[{right}]', end=' ')# this end will stop printing new line
    print('')
```

```python
dom=[]
for x in range(7):
    for y in range(x,7):
        dom.append((x,y))
dom
```

```python
dom[4][1] # row 4 col 1
dom[3][0]#
```

```python
pip_list=[]
for domini in dom:
    pip_list.append(domini[0]+domini[1])
pip_list
```

```python
list_compen=[y[0]+y[1] for y in dom ]
print(list_compen)
list_compen==pip_list
```

```python
def triangle(n):
    for x in range(1,n):
        for y in range(x,n):
            print('*', end=' ')
        print('')
triangle(6)
```

```python
def right_angle(n):
    for x in range(1,n):
        for y in range( x):
            print('*',end=' ')
        print()
right_angle(5)
```

```python
def triangle(n):
    for x in range(1,n):
        for y in range(x,n):
            print('*', end=' ')
        print('')
triangle(6)
```

```python
def right_angle(n):
    for x in range(1,n):
        for y in range( x):
            print('*',end=' ')
        print()
right_angle(5)
```

```python
def reverse_triangle(n):
   i=1
   j=i
   while i < n:
       while j < n:
           if i==n:
               print('*',end=' ')
           else:
               print('- ',end=' ')
           j+=1
       i+=1


       print()
reverse_triangle(5)
```

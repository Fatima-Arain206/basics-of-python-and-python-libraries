## .(number)f = round the data
## :(number) = allocate that many spaces
## :03 = allocate and pad that many spaces
 ## :< = left justify >> alignment for spaces
 ## :> = right justify
 ## :^ = center alignment
 ## :+ = use a plus sign to indicate positive value
 ## :=  = place sign to left most value
 ## :=  insert a space before positive value
## :, = comma separator



```python
price1 = 23.569086
price2 = -67.569073286
price3 = 93.569086943872
x=34
# to use specifier in placeholder use a colon :
print(f" price 1 is : {price1:.2f}") # round of 2 after . there should be 2 digit
print(f" price 2 is : {price2:.2f}")# f i for floating point
print(f" price 3 is : {price3:.0f}") # wants no gigt after .
# print(f" price 1 is : {x:.2f}")


```

```python
# now allocate the space we use a number after colon
print(f" price 1 is : {price1:10}") # it prints 2 spaces because there are 7 char in number
print(f" price 2 is : {price2:10}")# not any space
print(f" price 3 is : {price3:10}")# no space
a= 2
a2=-4.45
a3=6.6742
print(f" price 1 is : {a:10}")# 9 spaces
print(f" price 1 is : {a2:10}")# 6 spaces
print(f" price 1 is : {a3:10}")# 4

```

```python
# 0 pad use 0 before number
print(f" price 1 is : {a:02}")
print(f" price 2 is : {a2:06}")
print(f" price 3 is : {a3:010}")
```

```python
#left justify
print(f" price 1 is : {a:<10},") # this will print space to left side as we can see comma
print(f" price 1 is : {a2:<10},")
print(f" price 1 is : {a3:<10},")
f ="6.6742"
print(len(f))
print(f[1]) # digit is also a char
```

```python
# right justify
print(f" price 1 is :{a:>10},")# print 10 spaces from right
print(f" price 1 is :{a2:>10},")
print(f" price 1 is :{a3:>10},")

```

```python
# center align
print(f" price 1 is : {a:^4},")
print(f" price 1 is : {a2:^7},")
print(f" price 1 is : {a3:^5},")
```

```python
# positive
# it shows if the value is postive
print(f" price 1 is : {a:+},")
print(f" price 1 is : {a2:+},")# value is negative
print(f" price 1 is : {a3:+},")
```

```python
# space
print(f" price 1 is : {a: },")
print(f" price 1 is : {a2: },")# only one spce
print(f" price 1 is : {a3:},")
```

```python
# comma
a=30057893.4568
a2=903754384175.83
a3=7298195917.872
print(f" price 1 is : {a: ,}")# this is use for large values
print(f" price 1 is : {a2:,}")
print(f" price 1 is : {a3:,}")
```

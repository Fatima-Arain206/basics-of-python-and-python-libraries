```python

    
# Is dictionary mein har key-value ki data type alag hai
fatima_profile = {
    "name": "Fatima",          # Key: String, Value: String
    "semester": 4,             # Key: String, Value: Integer
    "gpa": 3.8,                # Key: String, Value: Float
    "is_active": True,         # Key: String, Value: Boolean (True/False)
    "skills": ["Python", "C++"] # Key: String, Value: List (List bhi daal di!)
}

print(fatima_profile["C++"]) # Output: ['Python', 'C++']
```

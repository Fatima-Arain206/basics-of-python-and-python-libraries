```python
import math
print(math.pi)# constant value of pi
# in e exoneration constant
print(math.e)
sq = math.sqrt(9)# this will return square root of 8
print(sq)
x = 8.912
# there is ceiling function it rounds floating point number up
print(math.ceil(x)) # now this is 9
# difference between round and ceil
print(round(x))# this simply rounds the digits

# floor it always round number number down
print(math.floor(8.912)) # it rounds to down
```

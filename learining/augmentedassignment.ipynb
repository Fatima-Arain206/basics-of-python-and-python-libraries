```python
friends = 5
friends =friends+1
print(friends)
# this is the simple assignment
```

```python
# augment
friends += 1 # for add
print(friends)
```

```python
#friends = friends-2
print(friends)
#subtraction
friends -= 2
print(friends)
# running again and again only this cell it will result as
# first run
# 7
#5
#again
# 5 3
# again
# 3 1 and going on
```

```python
# multiplication
#friends = friends *2
friends *= 2# 3*2 =6
print(friends)
```

```python
# division
# friends = friends/2 == 6/2 = 3.0
friends //=2 # friend /=2
print(friends)# 3.0
```

```python
# modulo
# friends = friends%2 == 1
friends %=2 # 1
print(friends)
```
